```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.sans-serif'] = ['Times New Roman'] 
plt.rcParams['axes.unicode_minus'] = False 
```

# 1 Loss and Accuracy

```python
train_loss = [i for i in range(50)]
val_loss = [i+1 for i in range(50)]
train_acc = [i for i in range(50)]
val_acc= [i+1 for i in range(50)]
epochs = [i for i in range(50)]

method = 'X method' #method name
```

```python
plt.figure(figsize=(6, 4.5))
plt.grid(linestyle="-")  
ax = plt.gca()
plt.plot(epochs, train_loss, color="#0DAB76", label="Train Loss", linewidth=1.5)
plt.plot(epochs, val_loss, color="#F487B6", label="Validation Loss", linewidth=1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Loss", fontsize=12)
plt.xlabel("#Epochs", fontsize=12)
plt.xlim(0, 48) # range of x
plt.ylim(0, 50) # range of y

plt.legend(loc=0, numpoints=1)
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=12) 

plt.title(method + ' Loss')

plt.savefig('./' + method + '_loss.pdf', format='pdf')  # save as pdf, then screen capture
plt.show()
```

```python
plt.figure(figsize=(6, 4.5))
plt.grid(linestyle="-")  
ax = plt.gca()
plt.plot(epochs, train_acc, color="#0DAB76", label="Train Accuracy", linewidth=1.5)
plt.plot(epochs, val_acc, color="#F487B6", label="Validation Accuracy", linewidth=1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Accuracy", fontsize=12)
plt.xlabel("#Epochs", fontsize=12)
plt.xlim(0, 48) # range of x
plt.ylim(0, 50) # range of y

plt.legend(loc=0, numpoints=1)
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=12) 

plt.title(method + ' Accuracy')

plt.savefig('./'+method+'_acc.pdf', format='pdf')  # save as pdf, then screen capture
plt.show()
```

## Combine two figures

```python
# Create a new figure with a specified size
plt.figure(figsize=(12, 4.5))

# First subplot: Loss
plt.subplot(1, 2, 1)
plt.grid(linestyle="-")  
plt.plot(epochs, train_loss, color="#0DAB76", label="Train Loss", linewidth=1.5)
plt.plot(epochs, val_loss, color="#F487B6", label="Validation Loss", linewidth=1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Loss", fontsize=12)
plt.xlabel("#Epochs", fontsize=12)
plt.xlim(0, 48) # range of x
plt.ylim(0, 50) # range of y
plt.legend(loc=0, numpoints=1)
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=12) 
plt.title(method + ' Loss')

# Second subplot: Accuracy
plt.subplot(1, 2, 2)
plt.grid(linestyle="-")  
plt.plot(epochs, train_acc, color="#0DAB76", label="Train Accuracy", linewidth=1.5)
plt.plot(epochs, val_acc, color="#F487B6", label="Validation Accuracy", linewidth=1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Accuracy", fontsize=12)
plt.xlabel("#Epochs", fontsize=12)
plt.xlim(0, 48) # range of x
plt.ylim(0, 50) # range of y
plt.legend(loc=0, numpoints=1)
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=12) 
plt.title(method + ' Accuracy')

plt.savefig('./'+method+'_combine.pdf', dpi=300, format='pdf')  # save as pdf, then screen capture
plt.show()
```
